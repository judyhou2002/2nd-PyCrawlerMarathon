## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime,  timedelta
import time
import pandas as pd
```

```python
# 打開瀏覽器

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get('https://www.ettoday.net/news/news-list.htm')
```

```python
#取得昨天的時間

# datetime.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)
#https://docs.python.org/3/library/datetime.html?highlight=timedelta#datetime.timedelta

#strftime 將 datetime object 轉換成String，strftime(format)
#strptime 將 datetime string 轉換成object，strptime(date_string, format)
#https://docs.python.org/3/library/datetime.html#strftime-strptime-behavior

yesterday = datetime.today() - timedelta(days = 1)
yesterday = yesterday.strftime( "%Y/%m/%d")
print( "昨天日期：" , yesterday)
```

```python
# 每個兩秒鐘自動往下滑
while True:
    browser.execute_script('window.scrollTo(0, 100000);')
    time.sleep(2)
    
    html_source = browser.page_source #網頁原始碼
    soup = BeautifulSoup(html_source, 'html5lib')
    print(soup.find_all(class_='part_list_2')[-1].find_all('h3')[-1].find(class_='date').text)
    
    if yesterday in soup.find_all(class_='part_list_2')[-1].find_all('h3')[-1].find(class_='date').text:
        print('==== STOP ====')
        break
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html5lib')

for d in soup.find(class_= 'part_list_2').find_all('h3'):
    if yesterday in d.find(class_='date').text:
        break
    print(d.find(class_= 'date').text, d.find_all('a')[-1].text)
```

```python
news = []
for d in soup.find(class_= 'part_list_2').find_all('h3'):
    if yesterday in d.find(class_='date').text:
        break
    news.append({
        '時間' : d.find(class_='date').text,
        '標題': d.find('a').text
    })
    
pd.DataFrame(news)
```

```python
# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
two_hours_ago_time = datetime.now() - timedelta(hours = 2)
two_hours_ago_time = two_hours_ago_time.strftime('%Y/%m/%d %H:%M') #轉換成str

#print(two_hours_ago_time)

news = []
for d in soup.find(class_= 'part_list_2').find_all('h3'):
    if d.find(class_='date').text < two_hours_ago_time:
        break
    news.append({
        '時間' : d.find(class_='date').text, 
        '標題': d.find('a').text
    })
    
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器

browser = webdriver.Chrome(executable_path = 'chromedriver')
browser.get('https://www.ettoday.net/news/news-list.htm')
```

```python
date_time = datetime.now() - timedelta(days = 3)
date_time_month = date_time.strftime('%#m')   # %m顯示01-12，%#m顯示1-12，去除前面的0
date_time_day = date_time.strftime('%#d')         # %d顯示01-31，%#d顯示1-31，去除前面的0

print('三天前日期:', date_time ) 
print('三天前日期(月):',date_time_month)
print('三天前日期(日):',date_time_day)

date_time_15 = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H') 
date_time_17 = datetime.strptime(date_time.strftime('%Y/%m/%d') + ' 17' , '%Y/%m/%d %H') 
print("取得昨天下午三點：", date_time_15)
print("取得昨天下午五點：", date_time_17)
```

```python
# 選擇且點選日期
from selenium.webdriver.support.ui import Select

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_time_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day)
browser.find_element_by_id('button').click()
```

```python
# 利用 selenium 持續執行下滑的動作，直到取得要求時間內的新聞，表示取得完畢
while True:
    browser.execute_script('window.scrollTo(0, 100000);')
    time.sleep(2)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html5lib')
    
    new_date = soup.find(class_='part_list_2').find_all('h3')[-1].find(class_='date').text
    print(new_date)
    
    if datetime.strptime(new_date, '%Y/%m/%d %H:%M') < date_time_15:  #轉換new_date格式從str →datetime
        print('===STOP===')
        break
```

```python
news = []
for d in soup.find(class_='part_list_2').find_all('h3'):
    if datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') < date_time_15:  #轉換date格式從str →datetime
        break
        
    elif datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') < date_time_17:
        news.append({
            '時間' : d.find(class_='date').text,
            '標題' : d.find('a').text
        })

pd.DataFrame(news)
```

```python
# 關閉瀏覽器

browser.close()
```

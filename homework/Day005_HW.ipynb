## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
data = json.loads(response)

print(len(data))

data1 = data[0]
print("Value : %s" %  data1.keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data:
    print(d['title'],d['createdAt'],d['commentCount'],d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

data_commentcount  = 0
data_likecount = 0
for d in data:
    data_commentcount += d['commentCount']
    data_likecount += d['likeCount']
    
average_data_commentcount = data_commentcount / len(data)
average_data_likecount = data_likecount / len(data)
print("熱門文章平均留言人數：", average_data_commentcount )
print("熱門文章平均按讚人數：", average_data_likecount)


r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2 = r2.text
data2 = json.loads(response2)

data2_commentcount = 0
data2_likecount  = 0
for d2 in data2:
    
    data2_commentcount += d2['commentCount']
    data2_likecount += d2['likeCount']

average_data2_commentcount = data2_commentcount / len(data2)
average_data2_likecount = data2_likecount / len(data2)
print("非熱門文章平均留言人數：", average_data2_commentcount )
print("非熱門文章平均按讚人數：", average_data2_likecount)
```

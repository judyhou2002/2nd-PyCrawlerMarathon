```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
import json
data = json.loads(response)

print(len(data))

data1 = data[0]
print("Value : %s" %  data1.keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d['title'],d['createdAt'],d['commentCount'],d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

data_commentcount  = 0
data_likecount = 0
for d in data:
    data_commentcount += d['commentCount']
    data_likecount += d['likeCount']
    
average_data_commentcount = data_commentcount / len(data)
average_data_likecount = data_likecount / len(data)
print("ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ï¼š", average_data_commentcount )
print("ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸ï¼š", average_data_likecount)


r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2 = r2.text
data2 = json.loads(response2)

data2_commentcount = 0
data2_likecount  = 0
for d2 in data2:
    
    data2_commentcount += d2['commentCount']
    data2_likecount += d2['likeCount']

average_data2_commentcount = data2_commentcount / len(data2)
average_data2_likecount = data2_likecount / len(data2)
print("éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ï¼š", average_data2_commentcount )
print("éç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸ï¼š", average_data2_likecount)
```

### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
res = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all("a")[-1].text
    tag = d.find(class_='tag').text
    
    #dict.setdefault(key[, default]) 如果 key 在 dict 中，回傳 value 值，反之，將 key:default 加入 dict 之中
    res.setdefault(tag,[])
    
    res[tag].append({
        'title' : title,
        'date' : date
    })
   
print(res)
```

### ③ 哪一個類別的文章最多

```python
data = []
for i in res:
    dic = {}
    dic['title'] = i
    dic['count']=len(res[i])
    data.append(dic)
    print(i, len(res[i]))
    
print (sorted(data, key = lambda k:k['count'], reverse=True))
'''
or
data.sort(key=lambda k:k['count'], reverse=True)
print(data)
'''
```

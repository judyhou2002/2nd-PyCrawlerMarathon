### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
import requests

headers = {
           'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36 OPR/68.0.3618.104 (Edition avira-2)'
          }
r = requests.get('http://www.dianping.com/', headers = headers)
response = r.text
response
```
